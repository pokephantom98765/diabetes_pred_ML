```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
data =  pd.read_csv('/content/diabetes.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['Outcome'].value_counts()
```

```python
Diabetes = data[data['Outcome']==0]
non_Diabetes =  data[data['Outcome']==1]
```

```python
Diabetes.head()
```

```python
non_Diabetes.head()
```

```python
Diabetes.shape
```

```python
non_Diabetes.shape
```

```python
Diabetes_sample = Diabetes.sample(n=268)
```

```python
new_data = pd.concat([Diabetes_sample,non_Diabetes],axis=0)
```

```python
new_data.head()
new_data.shape
```

```python
new_data.head()
```

```python
new_data['Outcome'].value_counts()
```

```python
new_data.describe()
```

```python
new_data.groupby('Outcome').mean()
```

```python
X= new_data.drop(columns='Outcome',axis=1)
Y= new_data['Outcome']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X = scaler.transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python

X_train_pred =classifier.predict(X_train)
train_acc = accuracy_score(X_train_pred,Y_train)
```

```python

X_test_pred =classifier.predict(X_test)
test_acc = accuracy_score(X_test_pred,Y_test)
```
